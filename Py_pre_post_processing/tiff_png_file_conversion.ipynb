```python
from PIL import Image
import os

# Paths
input_folder = r'D:\DeepLearning\HSVWesternDiagnosticMethods\CZ'
output_folder = r'D:\DeepLearning\HSVWesternDiagnosticMethods\CZ'

# Create output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Process each TIFF file in the input folder
for file_name in os.listdir(input_folder):
    if file_name.lower().endswith('.tiff') or file_name.lower().endswith('.tif'):
        # Open TIFF image
        tiff_path = os.path.join(input_folder, file_name)
        with Image.open(tiff_path) as img:
            # Convert to PNG and save
            png_path = os.path.join(output_folder, os.path.splitext(file_name)[0] + '.png')
            img.save(png_path, 'PNG')

print('Conversion completed!')
```
